```python
import pandas
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from datetime import datetime
from matplotlib import rcParams

def StringTimestamp():
    return datetime.now().strftime("%Y%m%d_%H%M%S_")

data_directory = 'C:/Users/rbk20xqk/Documents/google_backup/data/lorenz_equation/'
```

```python
lorenz_data = pandas.read_csv(data_directory + "lorenz_sequence.csv", header=None)
```

```python
lorenz_data
```

```python
plt.rcParams['legend.fontsize'] = 10

fig = plt.figure(figsize=(15,10))
ax = fig.gca(projection='3d')


ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
ax.zaxis.set_tick_params(labelsize=15)


ax.set_xlabel('x(t)', size = 20)
ax.set_ylabel('y(t)', size = 20)
ax.set_zlabel('z(t)', size = 20)

plt.plot(lorenz_data[1], lorenz_data[2], lorenz_data[3], linewidth=0.3)

plt.savefig(data_directory + "lorenz_attractor.pdf", format = 'pdf', bbox_inches='tight')
```

```python
bifurcation_data = pandas.read_csv(data_directory + "lorenz_bifurcation.csv", header=None)
```

```python
bifurcation_data
```

```python
plt.rcParams['legend.fontsize'] = 10

fig = plt.figure(figsize=(15,10))
ax = fig.gca()

ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)

ax.set_xlabel("'r'-parameter", size = 20)
ax.set_ylabel('x(t) local maxima', size = 20)

plt.scatter(bifurcation_data[1], bifurcation_data[2], marker='.', s=0.5)

plt.savefig(data_directory + StringTimestamp() + "lorenz_bifurcation.pdf", format = 'pdf', bbox_inches='tight')
```
